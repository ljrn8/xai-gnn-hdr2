```python
from pathlib import Path
from GNN_training_utils import *
from models import *
import os

from pathlib import Path
import os

def get_data(dataset_path):
    dataset_path = Path(dataset_path)
    graphs = {
        file.name: openpkl(file)
        for file in dataset_path.iterdir()
        if file.is_file()
    }
    models = {
        file.name: openpkl(file)
        for file in (dataset_path / "models").iterdir()
        if file.is_file()
    }

    return graphs, models


graphs, models = get_data(Path('output') / 'inductive_graph_tasks' / 'mutag')
```
