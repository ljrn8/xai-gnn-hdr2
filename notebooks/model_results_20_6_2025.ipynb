```python
import stable_gnn_xai
from stable_gnn_xai.util import openpkl, savepkl
from stable_gnn_xai.training.train import evaluate
from stable_gnn_xai.interfaces import ModelEvaluation
```

```python
import os
from pathlib import Path
import sys


runs_dir = Path('../output/runs')
for model_pkl in os.listdir(runs_dir):
    model_directory = runs_dir / model_pkl

    for model_file in os.listdir(model_directory):
        print(f'{model_pkl} on {model_file}')
        if not model_file.endswith(".pkl"):
            continue

        model_run = openpkl(model_directory / model_file)
        test_graphs = [g for g in openpkl(Path('..') / model_run.dataset_root) if g.test_mask == 1]
        print(
            f"Evaluating model {model_file} on test set of dataset {model_run.dataset_root}"
        )
        y_scores, metrics = evaluate(model_run.model, test_graphs)
        print('\n\n')
```
