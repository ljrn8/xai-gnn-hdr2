```python
import matplotlib.pyplot as plt
import numpy as np


with open('../Data/processed/ba2_graphs.pkl', 'rb') as f:
    graphs = pickle.load(f)

test_graphs = [g for g in graphs if g.test_mask == 1]
print(np.unique([g.y for g in test_graphs], return_counts=True))


proportions = [
    sum(g.edge_mask) / len(g.edge_mask) for g in graphs
]

print(proportions[0])
plt.hist(proportions, bins=100)
plt.show()

```
